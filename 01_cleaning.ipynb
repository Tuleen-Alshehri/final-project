# **Phase-1**

**In the first step, I downloaded the libraries I needed.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

**After that, I started exploring the data to figure out how to handle it.**

```python
#Since we want to clean the data all we need is the training file.
#step-1
df = pd.read_csv('train.csv')
print(df.head())
#step-2
print("\n\nRows:", df.shape[0])
print("Columns:", df.shape[1])
#step-3
print(df.describe())
print("info:",df.info())
#step-4
print("\n\n",df.isnull().sum())
```

**We've finished exploring the data now we start cleaning it up.**

```python
#step-5
def clean_data(df):
 df['Age'] = df['Age'].fillna(df['Age'].median())
 df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

 #We drop this column because most of its values ​​are missing.
 df = df.drop(columns=['Cabin'])

 #step-6
 print(df.duplicated().sum())
 #There are no duplicates so we don't need to drop anything.

 #step-7(outliers)
 #1-age

 Q1 = df["Age"].quantile(0.25)
 Q3 = df["Age"].quantile(0.75)

 IQR = Q3 - Q1

 lower = Q1 - 1.5 * IQR
 upper = Q3 + 1.5 * IQR

 df["Age"] = df["Age"].clip(lower, upper)

#2-fare
 Q1 = df["Fare"].quantile(0.25)
 Q3 = df["Fare"].quantile(0.75)

 IQR = Q3 - Q1

 lower = Q1 - 1.5 * IQR
 upper = Q3 + 1.5 * IQR

 df["Fare"] = df["Fare"].clip(lower, upper)
 return df

# step-8
percent_99 = df['Fare'].quantile(0.99)
df = df[df['Fare'] <= percent_99]
```

```python
#step-9
df = clean_data(df)
#check-1
assert df['Age'].isnull().sum() == 0
#check-2
assert (df['Fare'] >= 0).all()
#check-3
assert df.shape[1] == 11
```
